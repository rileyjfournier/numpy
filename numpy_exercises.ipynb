#### 17. Given a 1D array, negate all elements which are between 3 and 8, in place. (★☆☆)

#### 25. Create random vector of size 10 and replace the maximum value by 0 (★★☆)

````python
a = np.random.rand(10)
amax = a.max()
i = 0
for num in a:
    if num == amax:
        a[i] = 0
    i += 1
    
print(a)
````

#### 27. Convert a float (32 bits) array into an integer (32 bits) in place?

````python
arr = np.array([1.,2.,3.,4.], dtype='float32')
print('Before:', arr, '| data type: ', arr.dtype)
print('CONVERTING!')
arr = arr.astype('int32')
print('After:', arr, '| data type: ', arr.dtype)
````

#### 28. Sort an array by the nth column? (★★☆)

````python
a = np.random.randint(1, 10, (3,3))

def sort_by_column_n(arr, n):
    b = arr[arr[:, n].argsort()]
    return b

sort_by_column_n(a, 0)
````

#### 30. Consider the vector \[1, 2, 3, 4, 5\], how to build a new vector with 3 consecutive zeros interleaved between each value? (★★★)

````python
arr = np.array([1,2,3,4,5])
z = 3
new_arr = np.zeros(len(arr) + (len(arr)-1)*(z))
new_arr[::z+1] = arr
print(new_arr)
````

#### 34. Get the n largest values of an array (★★★)

````python
a = np.array([2,0,4,5,7,2,9,6])

sorted_index_array = np.argsort(a)

sorted_array = a[sorted_index_array]

n = 1

print(sorted_array[-n:])
````
